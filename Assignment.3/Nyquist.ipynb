```python
s = control.TransferFunction.s

# G_1 = (20 * (s+8)) / (s * (s+2) * (s+4))
G_2 = 1 / (s**2 + 1)


w = np.logspace(-3,3,50000)
control.nyquist(G_2,w);
plt.grid(True)
plt.title('Nyquist Diagram')
plt.xlabel('Re(s)')
plt.ylabel('Im(s)')
plt.show()
```
